```python
df = pd.read_csv('../data/subtask_a_train.tsv', sep='\t')
print(df["sentence_type"].value_counts())
df.head(10)
# df.iloc[0]['expected_order']
```

```python
from pathlib import Path

# Get all compounds from the dataframe
compounds_in_df = set(df['compound'].unique())

# Get all directory names from train folder
train_path = Path('../data/train')
train_dirs = {d.name for d in train_path.iterdir() if d.is_dir()}

# Find compounds that don't have corresponding directories
missing_dirs = compounds_in_df - train_dirs

print("Compounds without corresponding directories:")
for compound in sorted(missing_dirs):
    print(f"- {compound}")
print(f"\nTotal missing: {len(missing_dirs)}")
```
